```python
import numpy as np
```

```python
print(np.array([1,2,3,4,5]))
```

```python
#（5）或者(5,)
zeros = np.zeros((5))
print(zeros)
```

```python
ones = np.ones((5))
print(ones)
```

```python
randoms = np.random.random((5))
print(randoms)
```

```python
#二维数组，2行5列
zeros = np.zeros((2, 5))
print(zeros)
```

```python
arange = np.arange(10)
print(arange)
```

```python
reshape = arange.reshape((2,5))
print(reshape)
```
